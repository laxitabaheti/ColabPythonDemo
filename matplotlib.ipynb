```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5, 6 ])
y = np.array([.10, 10,.20, 20, .30, 30, .40])
plt.plot(x,y, 'o-.m',ms =10,mfc='c')
plt.show()
```

- **ms** = marker size
- **mfc** = marker face color
- **format string ( fmt )** = marker|line|color
    1.   marker = 'o' , '+' , 'd' , 'D' , '1' , '2' , '3' , '4' , 'h' , 'p' , etc.
    2.  line style =  '-' ( solid line ) , '--' ( dashed line ) , ':' ( dotted line ) , '-.' ( dashed dotted line )
    3. color of the line = 'r' ( red ) , 'b' ( blue ) , 'k' ( black ) , 'm' ( magenta ) ,  'c' ( cyan ) , 'w' ( white ) , 'y' ( yellow ) , 'o' ( orange )

```python
import numpy as np
import matplotlib.pyplot as plt


font1 = {'family':'serif','color':'g','size':12}
font2 = {'family':'serif','color':'b','size':10}

x = np.array([0, 1, 2, 3])
y = np.array([4, 8, 1, 10])

plt.subplot(2,2,1)
plt.plot(x,y,'',marker='o',mec='pink',mfc='k',ms=5)
plt.title("Graph1",fontdict=font2,loc='left')
plt.xlabel("x", fontdict=font1)
plt.ylabel("y",fontdict= font1)



x = np.array([0, 5, 2, 8])
y = np.array([10, 20, 30, 40])

plt.subplot(2,2,2)
plt.plot(x,y,marker='o',mec='pink',mfc='k',ms=5)
plt.title("Graph2",fontdict=font2,loc='right')
plt.xlabel("x", fontdict=font1)
plt.ylabel("y", fontdict=font1)



x = np.array([0, 4, 2, 13])
y = np.array([10, 20, 30, 40])

plt.subplot(2,2,3)
plt.plot(x,y,marker='o',mec='pink',mfc='k',ms=5)
plt.title("Graph3",fontdict=font2,loc='left')
plt.xlabel("x", fontdict=font1)
plt.ylabel("y", fontdict=font1)



x = np.array([0, 1, 2, 3])
y = np.array([3, 1, 3, 2])

plt.subplot(2,2,4)
plt.plot(x,y,marker='o',mec='pink',mfc='k',ms=5)
plt.title("Graph4",fontdict=font2,loc='right')
plt.xlabel("x", fontdict=font1)
plt.ylabel("y", fontdict=font1)


plt.subplots_adjust(left=0.2,
                    bottom=0.2, 
                    right=0.9, 
                    top=.9, 
                    wspace=0.6, 
                    hspace=0.6)

plt.tight_layout()

plt.suptitle("Mapping",color='g',size=20)
plt.show()
```

```python
# To find the outer of two input arrays
import numpy as np


# Initialization of a 2x2 matrix
# as first input
a = np.arange(4)

# Initialization of an array as
# second input
b = np.arange(3)

# Outer of a & b
z = np.add.outer(b, a)

print("The outer of {0} & {1} is {2}".format(b,a,z))
```
